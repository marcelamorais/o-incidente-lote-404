# Pergunta 1: **Qual máquina causou o problema?**

Vamos verificar **status**, **alerta** e **histórico** das máquinas.

* `alerta = Pressão Crítica`

# Pergunta 2: **Por que a temperatura subiu tanto?**

Vamos analisar os **insumos**, especialmente a **Soda Cáustica**.

```python
df_insumos[["nome", "nivel", "st"]]
```

Agora filtrar apenas a Soda Cáustica:

```python
soda = df_insumos[df_insumos["nome"] == "Soda Cáustica"].iloc[0]
pd.DataFrame(soda["historico"])
```

Evolução clara:

* 95% → 80% → 65% → 50% → 35% → 28% → **22%**
* Status: **CRÍTICO**

**Lógica de processo industrial**:

* Pouca soda = reação química fraca
* Para compensar → **aumenta temperatura e pressão**
* Resultado → **papel queimado**

**Conclusão 2:**:

A temperatura subiu porque o **nível crítico de Soda Cáustica (22%) forçou o aumento térmico no Digestor**.
